## 4. Append to an Excel file

```python
# Create a small DataFrame with new order data to append
new_order_data = {'Row ID': [9995], 'Order ID': ['CA-2025-123456'], 'Order Date': ['2025-09-12'],
                  'Ship Date': ['2025-09-16'], 'Ship Mode': ['Standard Class'], 'Customer ID': ['CH-12070'],
                  'Customer Name': ['Chris H'], 'Segment': ['Consumer'], 'Country': ['United States'],
                  'City': ['Los Angeles'], 'State': ['California'], 'Postal Code': [90001],
                  'Region': ['West'], 'Product ID': ['OFF-LA-10000001'], 'Category': ['Office Supplies'],
                  'Sub-Category': ['Labels'], 'Product Name': ['Colored Labels'], 'Sales': [15.50],
                  'Quantity': [1], 'Discount': [0.0], 'Profit': [7.05]}
new_orders_df = pd.DataFrame(new_order_data)
```

## 5.  Read a Excel Chunk-by-chunk

```python
# File path 
file_path = "Orders.xlsx"

# Get total number of rows (without loading entire file)
total_rows = pd.read_excel(file_path, engine="openpyxl").shape[0]

# Define chunk size
chunk_size = 2000  # number of rows per chunk

# Read in chunks
for start_row in range(0, total_rows, chunk_size):
    # Read a chunk
    chunk = pd.read_excel(
        file_path,
        engine="openpyxl",
        skiprows=range(1, start_row + 1),  # skip rows except header
        nrows=chunk_size
    )
    
    print(f"Processing rows {start_row + 1} to {start_row + len(chunk)}...")
    print(chunk.head())  
```

# TASK-3: Handling JSON

## 1. Write the output of a Python Code into a JSON Format/File

```python
import json 
```

```python
# Extracting high-profit orders in a specific region from Orders.xlsx
high_profit_west_df = df[(df['Region'] == 'West') & (df['Profit'] > 500)]
print(high_profit_west_df)
```

```python
# The 'orient' parameter controls the JSON format.
# 'records' creates a list of JSON objects, with each object being a row.
# The 'indent' parameter makes the file human-readable.
high_profit_west_df.to_json(
    'high_profit_west.json',
    orient='records',
    indent=4
)

print("High-profit orders in the West region have been written to high_profit_west.json")
```

```python
output_filename = 'high_profit_west.json' #Write the filtered DataFrame to a JSON file.
high_profit_west_df.to_json(
    output_filename,
    orient='records',
    indent=4
)
```

```python
print(f"Displaying the content of {output_filename}")
with open(output_filename, 'r') as json_file:
    # Read the data from the JSON file into a Python object (e.g., list of dictionaries).
    data = json.load(json_file)
    
    # Use json.dumps() with indent for a nicely formatted print to the console.
    print(json.dumps(data, indent=4))

print(f"\nHigh-profit orders in the West region have been written to and displayed from {output_filename}")
```

## 2.Parse a JSON file and search for a value

```python
import json

# Define the filename and the value you want to find.
filename = 'high_profit_west.json'
search_value = 'California'

# Read and parse the JSON file.
try:
    with open(filename, 'r') as file:
        # data is a list of dictionaries because of `orient='records'`
        data = json.load(file)
except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
    data = None
except json.JSONDecodeError:
    print(f"Error: The file '{filename}' is not a valid JSON file.")
    data = None

# Proceed only if the file was loaded successfully.
if data and isinstance(data, list): # Check if data is a list
    # The data variable is already the list of records.
    found_orders = []
    for order in data:
        if order.get('State') == search_value:
            found_orders.append(order)

    # Print the results.
    if found_orders:
        print(f"Found {len(found_orders)} high-profit order(s) in {search_value}:")
        for order in found_orders:
            # Adjust the print statement based on what keys are in your data.
            print(f" - Order ID: {order.get('Order ID')}, Profit: {order.get('Profit')}")
    else:
        print(f"No high-profit orders found in {search_value}.")
else:
    print("Failed to load or parse JSON file correctly.")
```
